Using Tensorflow for Deep Learning and Tensorboard to illustrate and save our data models

```python
from tensorflow.keras.callbacks import TensorBoard
```

```python
from datetime import datetime
```

The log directory indicates where our data models will be saved.

```python
log_directory = 'logs/fit'
```

```python
timestamp = datetime.now().strftime("%Y-%m-%d--%H:%M")
```

```python
log_directory = log_directory + '/' +timestamp
log_directory
```

```python
board = TensorBoard(log_dir = log_directory, 
                   histogram_freq=1,
                   write_graph = True,
                   write_images = True,
                   update_freq = "epoch",
                   )
```

```python
import pandas as pd
df = pd.read_csv("bank_note_data.csv")
```

```python
df[df.isnull()==False]
```

Partitioning the data and transforming the data (scaling and normalization of the data isn't needed because the numbers are small)

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('Class', axis =1), df['Class'], test_size = 0.3)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

Adding Dropout Layers and Early stopping in order to prevent overfitting

```python
model = Sequential()
model.add(Dense(30, activation = "relu"))
model.add(Dropout(0.2))
model.add(Dense(15, activation = "relu"))
model.add(Dropout(0.2))
model.add(Dense(1, activation = "sigmoid"))
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor = "val_loss", mode = "min", verbose = 1, patience = 25)
```

```python
model.fit(x_train, y_train.values.reshape(960,1), epochs = 600, validation_data = (x_test, y_test.values.reshape(412,1)), verbose =1 ,callbacks=[early_stop, board])
```

Graphs and models are saved within the logs/fit folder. We can predict values and test the accuracy of our neural network as well

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
predictions = model.predict_classes(x_test)
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
#ingore warning
```

Our neural network worked perfectly with an accuracy rate of 100%.
