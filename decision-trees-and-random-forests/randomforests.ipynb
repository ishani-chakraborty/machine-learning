Training and evaluating a Random Forest Model to correctly predict whether a loan will be paid back by a person.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
loans = pd.read_csv("loan_data.csv")
```

Understanding information about the features of the dataset.

```python
loans.info()
```

```python
loans.describe()
```

Data visualization

```python
sns.distplot(loans[loans['credit.policy']==0]['fico'], kde = False, bins = 40, color = "PURPLE")
sns.distplot(loans[loans['credit.policy']==1]['fico'], kde = False, bins = 40, color = "YELLOW")
```

```python
sns.countplot(x = 'purpose', hue = 'not.fully.paid', data = loans)
```

```python
sns.lmplot(x = "fico", y = "int.rate", hue = "credit.policy" , col ='not.fully.paid', data = loans, legend = True )
```

Transforming Categorical Variables into Dummy Training Variables 

```python
loans.head()
```

The 'purpose' column is identified as categorical variable in need of transformation

```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(loans, columns = cat_feats, drop_first = True)
```

```python
final_data
```

Partitioning the data and train the Random Forest

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(final_data.drop('not.fully.paid', axis = 1), final_data['not.fully.paid'], test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 200)
```

```python
rfc.fit(x_train,y_train)
```

Using the trained model to predict whether loan will be paid back, and evaluating the accuracy of our model.

```python
predict = rfc.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, predict))
print(classification_report(y_test, predict))
```

The Random Forest Model performed better than the decision tree, almost achieving an 80% accuracy rate.
